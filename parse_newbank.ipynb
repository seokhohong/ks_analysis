```python
import pandas as pd
import numpy as np
```

```python
with open('eu_bank.txt') as f:
    lines = f.readlines()
```

```python
class Game:
    def __init__(self, line):
        line = line.strip()
        print(line)
        components = line.split(' / ')[1:]
        #print(components)
        self.timestamp = components[0]
        self.win_prob = components[-1]
        self.roles = []
        self.mmrs = []
        for i in range(1, len(components) - 2, 2):
            self.roles.append(components[i])
            self.mmrs.append(int(components[i + 1]))
    def has_role(role):
        return role in self.roles
    def has_role_min_mmr(role, mmr):
        for i, r in enumerate(self.roles):
            if r == role:
                if self.mmrs[i] >= mmr:
                    return True
        return False
    def surv_mmrs():
        mmrs = []
        for m, r in zip(self.mmrs, self.roles):
            if not r in ['Kerrigan', 'Zagara', 'Thakras', 'Niadra', 'Dehaka']:
                mmrs.append(m)
    def kerri_mmrs():
        mmrs = []
        for m, r in zip(self.mmrs, self.roles):
            if r in ['Kerrigan', 'Zagara', 'Thakras', 'Niadra', 'Dehaka']:
                mmrs.append(m) 
        return mmrs
    def kerri_win_prob():
        return 1 - self.win_prob
    def surv_win_prob():
        return self.win_prob
            
games = [Game(line) for line in lines if len(line) > 0]
```

```python
mmrs = []
for game in games:
    mmrs.extend(game.mmrs)
    
np.mean(mmrs)
```
